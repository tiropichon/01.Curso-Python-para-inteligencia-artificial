Implementa los siguientes métodos especiales para la clase Lista:
- `__len__`: devuelve el tamaño de `lista` de la clase `Lista`.
- `__add__`: devuelve un objeto de tipo `Lista` con atributo `lista` con todos los elementos de las dos listas.
- `__mul__`: devuelve un objeto de tipo `Lista` con atributo `lista` con la lista `a` multiplicado por el tamaño de la lista `b`.
- `__lt__`: compara los tamaños de las dos listas.


````python

# Pablo Cóceres Eguía
#### Ejercicio 1F (MEDIO)

import random

class Lista():

    def __init__(self, lista):
        self.lista = lista

    def __str__(self):
        return str(self.lista)

    def __len__(self):
        return len(self.lista)

    def __add__(self, lista2):
        return self.lista + lista2.lista
    
    def __mul__(self, lista2):
        return self.lista * lista2

    def __lt__(self, lista2):
        return len(self.lista) < len(lista2)

randlist_a = random.sample(range(50), random.randint(1, 10))
list_a = Lista(randlist_a)
randlist_b = random.sample(range(50), random.randint(1, 10))
list_b = Lista(randlist_b)

print("Primera lista: ", list_a)
print("Segunda lista: ", list_b)
print("El tamaño de la primera lista es: ", len(list_a))
print("El tamaño de la segunda lista es: ",len(list_b))
print("Concatenación de las dos listas: ", list_a + list_b)
print("Primera lista multiplicada por el largo de la segunda: ", list_a * len(list_b))
print("El largo de la primera lista es menor que el de segunda lista?", list_a < list_b)

````
